From the above info we'll have to change the Date column from object data type to datetime.

```python
#Changing the bitcoin Date column's data type from object to date/time.
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
#displaying results
bitcoin_data.info()
```

```python
#Changing the ethereum Date column's data type from object to date/time.
ethereum_data['Date'] = pd.to_datetime(ethereum_data['Date'])
#displaying results
ethereum_data.info()
```

```python
#Changing the tether Date column's data type from object to date/time.
tether_data['Date'] = pd.to_datetime(tether_data['Date'])
#displaying the results
tether_data.info()
```

```python
#Ensure uniformity by changing the column names to lower case, stripping of the whitespaces,removing periods, 
#this ensures good readerbility plus every possible naming issue is covered.
#This is to be applied to the bitcoin, tether and ethereum datasets.
#Bitcoin dataset

bitcoin_data.columns = bitcoin_data.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
bitcoin_data.columns
```

```python
#Ethereum dataset
ethereum_data.columns = ethereum_data.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
ethereum_data.columns
```

```python
#Tether dataset
tether_data.columns = tether_data.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
tether_data.columns
```

```python
#Renaming the columns in both our datasets to differentiate between the bitcoin and ethereum columns by adding an 'eth' suffix for ethereum
#and a 'bit' for bitcoin.
bitcoin_data2 = bitcoin_data.add_suffix('bit')
#previewing results
bitcoin_data2.head()
```

```python
#repeat the process for ethereum dataset adding "eth"

ethereum_data2 = ethereum_data.add_suffix('eth')
#previewing results
ethereum_data2.head()
```

```python
#repeat the process for tether dataset adding "eth"

tether_data2 = tether_data.add_suffix('tet')
#previewing results
tether_data2.head()
```

```python
#renaming the 'dateeth' column to 'datebit' to be able to merge the two datasets since the data was collected in the same period.

ethereum_data3 = ethereum_data2.rename(columns={'dateeth': 'datebit'}, inplace=True)
```

```python
#renaming the 'dateeth' column to 'datebit' to be able to merge the two datasets since the data was collected in the same period
tether_data3 = tether_data2.rename(columns={'datetet': 'datebit'}, inplace=True)
```

```python
tether_data2.head()
```

```python
tether_data2.info()
```

#Merging the datasets

```python
# creating a new data set "crypto_data" by combining the Bitcoin, Tether and Ethereum datasets and displaying it.
#crypto_data = pd.merge(bitcoin_data2, ethereum_data2, left_on='datebit', right_on='datebit')
#crypto_data

crypto_df = bitcoin_data2.merge(tether_data2,on='datebit').merge(ethereum_data2,on='datebit')
```

## 2018 Crash

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python

def PlottingFunctionThree(x,y1,y2,y3,name1,name2,name3, color1, color2, color3, title,xtitle,ytitle):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x= x, y= y1, name= name1, line=dict(color= color1, width=2)))
    fig.add_trace(go.Scatter(x= x, y= y2, name= name2, line=dict(color= color2, width=2)))
    fig.add_trace(go.Scatter(x= x, y= y3, name= name3, line=dict(color= color3, width=2)))
    fig.update_layout(title = title,
            xaxis_title=xtitle,
            yaxis_title= ytitle)
    fig.show()
```

```python
crashperiod = crypto_df[(crypto_df['datebit'] > '2017-09-01') & (crypto_df['datebit'] < '2018-12-31')]
PlottingFunctionThree(crashperiod.datebit, crashperiod.adj_closebit, crashperiod.adj_closeeth, crashperiod.adj_closetet, 'Bitcoin', 'Ethereum', 'Tether', 'orange', 'purple','green',
 '2018 Crash Performance', 'Date', 'Closing Price')
```

```python
precovidperiod = crypto_df[(crypto_df['datebit'] > '2017-09-01') & (crypto_df['datebit'] < '2018-12-31')]
PlottingFunctionThree(crashperiod.datebit, precovidperiod.adj_closebit, precovidperiod.adj_closeeth, precovidperiod.adj_closetet, 'Bitcoin', 'Ethereum', 'Tether', 'orange', 'purple', 'green',
 'Pre-Covid Performance', 'Date', 'Closing Price')
```
